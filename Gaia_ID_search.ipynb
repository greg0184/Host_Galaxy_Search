# Search Gaia DR3 for these stars

```python
#columns = 'source_id', 'ra', 'dec', 'parallax', 'distance_gspphot', 'l', 'b', 'phot_g_mean_mag', 'bp_rp', 'phot_bp_mean_mag', 'phot_rp_mean_mag', 'teff_gspphot', 'mh_gspphot'

query_base = """ SELECT
    source_id, 4.74057*pm*distance_gspphot/1000 AS v_tan,
    ra, dec, ref_epoch, distance_gspphot, parallax, phot_g_mean_mag, phot_rp_mean_mag, phot_bp_mean_mag,
    g_rp, bp_rp, ag_gspphot, ebpminrp_gspphot, pm, teff_gspphot, teff_gspphot_lower, teff_gspphot_upper,
    logg_gspphot, mh_gspphot
    FROM gaiadr3.gaia_source
    WHERE source_id = {id}
    """
```

```python
id = 2940856402123426176
query = query_base.format(id = id)    
job = Gaia.launch_job(query)    
results = job.get_results()
results
```
